## 7. Interpret K-Means Clustering Quality

### Metric Explanation

**Adjusted Rand Index (ARI):**
- Range: -1 to +1
- **+1**: Perfect agreement between clusterings
- **0**: Random labeling (baseline)
- **-1**: Perfect disagreement
- Interpretation: Measures whether K-Means recovered the same pairwise cluster structure as true labels

**Normalized Mutual Information (NMI):**
- Range: 0 to 1
- **1**: Perfect mutual agreement (same information content)
- **0**: Independent clusterings
- Interpretation: Measures how much information K-Means labels share with true labels

### Justification for k=3

The elbow curve shows:
- **k=2 to k=3**: Steep decline in inertia (significant improvement)
- **k=3 to k=4**: More gradual decline in inertia
- **k=4 onwards**: Marginal improvements

The "elbow" point at k=3 indicates the optimal balance between model complexity and fit quality. Additionally, the seeds dataset is known to contain 3 wheat varieties (Kama, Rosa, Canadian), so k=3 aligns with domain knowledge.
